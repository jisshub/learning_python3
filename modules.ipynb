# built in modules

```python
#importing built in module random ans access the choice()
import random
print(random.choice(['a','f','r','t']))
```

```python
#vreate an alias for random and v can refer to random module using rand
import random as rand
print(rand.randint(8,100))
```

```python
#importing parts of module using from 
from random import randint,choice
#.no need to specify module name again while using its functions
print(randint(1,100))
print(choice(['t','y','u','i']))
```

```python
#importing parts of module using from 
from random import randint as r,choice as c
print(r(1,100))
print(c(['t','y','u','i']))
```

```python
# Import the math module:

# Use math.sqrt  to find the square root of 15129 and save it to variable called answer:
from math import sqrt as s
answer=s(15129)
print(answer)
```

```python
#checking whether the argument passed is a keywords.use keyword module 
from keyword import iskeyword as ik
def contains_strings(*args):
    for new in args:
        if ik(new):
            return True
        return  False
contains_strings('def','d','e','ee')
```

```python
import pyfiglet
from termcolor import colored
text=input("Enter the message: ")
color=input("Enter the color: ")
colors_avail=['red', 'green', 'yellow', 'blue', 'magenta', 'cyan', 'white']
if color in colors_avail:
    message=pyfiglet.figlet_format(text)
    color_me=colored(message,color)
    print(color_me)
else:
    color='magenta'
    color_me=colored(message,color)
    print(color_me)
```

```python
from pyfiglet import figlet_format as ff
from termcolor import colored as cr
msg=input('enter text: ')
color=input('enter color: ')
def change_me(msg,color,*args):
    if color in args:
        message=ff(msg)
        final=cr(message,color)
        print(final)
    else:
        print(f"{color} not available")
        print(f"choose one from here: {args}")
colors_avail=['red', 'green', 'yellow', 'blue', 'magenta', 'cyan', 'white']
change_me(msg,color,*colors_avail)
    
```
